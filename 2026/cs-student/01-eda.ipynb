```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset: https://www.kaggle.com/datasets/nalisha/computer-science-students-performance

```python
path = r"C:\Users\Rudra\Desktop\kaggle\cs-student\cs_students.csv"
```

```python
df = pd.read_csv(path)
df.head(2)
```

```python
df.info()
```

```python
# unique value
for col in df.columns:
    print(f"{col} -> {df[col].nunique()}")
```

```python
# major and name not imp so del
df.drop(columns=["Major", "Name"], inplace=True)
```

```python
df.columns
```

```python
df.describe()
```

```python
df['GPA'].value_counts().plot.bar(title="CGPA Bar chart")
plt.show()
```

```python
plt.plot(df['Student ID'], df['Age'])
plt.title('Age Distribution by Student ID')
plt.xlabel('Student ID')
plt.ylabel('Age')
plt.show()
```

```python
sns.barplot(
    data=df.groupby(["Future Career", "Gender"]).size().reset_index(name='Count').sort_values(by="Future Career").head(12),
    x='Count', 
    y='Future Career', 
    hue='Gender'
)
plt.title("Career choice by gender")
plt.show()
```

```python
df["Future Career"].value_counts().head(12).plot.barh(title="Future Career count")
plt.show()
```

```python
df["Projects"].value_counts().head(12).plot.barh(title="Project count")
plt.show()
```

```python
df['Interested Domain'].value_counts().plot.barh(title="Interest count on the domain")
plt.show()
```

```python
def plot_pie(col):
    df[col].value_counts().plot.pie(autopct="%2.2f%%")
    plt.title(f"{col} Count Distribution")
    plt.show()
```

```python
for col in ["Gender", "Python", "SQL", "Java"]:
    plot_pie(col)
```
