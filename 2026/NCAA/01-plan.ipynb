# Basic

> Few terms that is need to understand if you are beginner just like me.
>
> *AI Generated 

**What is NCAA?**

**NCAA** = *National Collegiate Athletic Association*

It is the organization that runs sports competitions for **U.S. colleges and universities**.

Think of it like:

* FIFA → controls World Cup (soccer)
* NBA → professional basketball league
* **NCAA → college basketball tournaments**

Every year they organize a big tournament called **March Madness**.




**What is a Team?**

A **team** is a college basketball team.

Example:

* Duke University basketball team
* Kansas University basketball team

In the dataset:

* Each team has a **TeamID** (a number)
* Example: TeamID = 1104
* That number represents one specific college team

Important:
The same team plays every year, but their players change.




**What is a Seed?**

A **seed** = ranking number given before the tournament starts.

It shows how strong a team is expected to be.

Think of it like:

* Seed 1 → very strong team
* Seed 16 → weak team

Each region has seeds:
1, 2, 3, … up to 16

Example:
W01 = Region W, Seed 1 (very strong)
X16 = Region X, Seed 16 (underdog)

Why seeds matter:
Higher seed teams usually win more often.




**What is DayNum?**

DayNum = Day Number of the season.

Instead of real dates (Jan 10, Feb 2, etc.), they count days from the start of the season.

Example:

* DayNum 0 → first day of season
* DayNum 132 → Selection Sunday (seeds announced)
* DayNum 154 → Championship game

Why they use DayNum:
It makes it easier to compare seasons without worrying about calendar dates.




**What is the Tournament?**

After the regular season:

1. Best teams are selected
2. They are given seeds
3. They play knockout matches
4. Loser is eliminated
5. Winner continues
6. Final winner = National Champion

It’s like:
World Cup knockout stage.




**What is Regular Season?**

Before the tournament, teams play many games.

These games:

* Help determine how strong teams are
* Help decide tournament selection

This data is used to:

* Calculate team strength
* Train prediction models




**What is in the Dataset?**

The dataset gives:

* Which team played which team
* Who won
* Score
* Location (home/away/neutral)
* Seed
* Season year

Your job in Kaggle:
Predict the probability that Team A beats Team B in the tournament.




**Very Simple Summary**

- NCAA → Organization that runs college basketball
- Team → A college basketball team
- Season → One year of games
- DayNum → Day number inside that season
- Seed → Pre-tournament ranking (1 = strongest)
- Tournament → Knockout competition to decide champion
