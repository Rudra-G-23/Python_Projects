```python
import folium

# Center map around Egypt
egypt_center = [df["Customer_Lat"].mean(), df["Customer_Lon"].mean()]

m = folium.Map(location=egypt_center, zoom_start=6)

# Add small sample (don't plot all 100k)
sample_df = df.sample(1000)

for _, row in sample_df.iterrows():
    folium.CircleMarker(
        location=[row["Customer_Lat"], row["Customer_Lon"]],
        radius=2,
    ).add_to(m)

m
```

```python
city_name = "Cairo"
city_df = df[df["City"] == city_name]

city_center = [city_df["Customer_Lat"].mean(),
               city_df["Customer_Lon"].mean()]

m_city = folium.Map(location=city_center, zoom_start=12)

sample_city = city_df.sample(1000)

for _, row in sample_city.iterrows():
    folium.CircleMarker(
        location=[row["Customer_Lat"], row["Customer_Lon"]],
        radius=3,
    ).add_to(m_city)

m_city
```
