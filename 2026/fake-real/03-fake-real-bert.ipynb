
**Traditional:**

- TF-IDF → frequency based
- Word2Vec → static embeddings

**BERT:**
- It understand the context means Context-aware embeddings
- Understands sentence meaning
- Bidirectional (looks left + right)

Ex:
- Sentence 1:The bank approved the loan
- Sentence 2:He sat on the river bank

Word "bank" gets different vectors in each sentence
TF-IDF & Word2Vec cannot do this properly.

```python
path = r"C:\Users\Rudra\Desktop\kaggle\fake-real\fake_news_dataset_4000_rows.csv"
df = pd.read_csv(path)
df.sample(2)
```

```python
df["combined"] = df["title"] + " " + df["text"]
```

# Train test split

```python
train_df, test_df = train_test_split(
    df[["combined", "label"]],
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)
```

```python
print(train_df.shape, test_df.shape)
```

# Convert to huggingface datasets

```python
train_dataset =  Dataset.from_pandas(train_df.reset_index(drop=True))
test_dataset = (Dataset.from_pandas(test_df.reset_index(drop=True)))
```
