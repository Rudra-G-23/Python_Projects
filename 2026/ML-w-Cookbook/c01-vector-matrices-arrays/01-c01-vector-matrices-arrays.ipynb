# Creating a Sparse Matrix

```python
from scipy import sparse
```

```python
mtx = np.array([
    [0, 2],
    [3, 0],
    [0, 567]
])

mtx
```

```python
mtx_sparse = sparse.csr_matrix(mtx)
```

```python
mtx_sparse
```

```python
print(mtx_sparse)
```

# 1.18 Adding and Subtracting Matrices
