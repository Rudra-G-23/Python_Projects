<a href="#table-of-contents" style="font-family: Arial, sans-serif; font-size: 15px; background-color: #7b2cbf; color: #ffffff; padding: 7px 15px; text-decoration: none; border-radius: 15px;">

```python
work_type_df = df.melt(
    id_vars=['Age', 'Gender'],
    value_vars=['Social_Media_Usage_Hours', 'Productivity_App_Usage_Hours', 'Gaming_App_Usage_Hours'],
    var_name='App_Usage_Type',
    value_name='Usage_Hours',
)

work_type_df.sample(3)
```

```python
fig = px.histogram(
    work_type_df,
    x='Usage_Hours',
    color='App_Usage_Type',
    barmode='group',
    
)

fig.update_layout(
    title='App usages Hours ',
    bargap=0.3
)

fig.show()
```

### <div style="text-align:left; border-radius:15px; padding:20px; color:Black; margin:0; font-size:100%; font-family:Arial; background-color:#48cae4; overflow:hidden"><b>‚ö° 6. Distribution of App Usage Hours by Type</b>
<a href="#table-of-contents" style="font-family: Arial, sans-serif; font-size: 15px; background-color: #7b2cbf; color: #ffffff; padding: 7px 15px; text-decoration: none; border-radius: 15px;">

```python
app_usage_summary = work_type_df.groupby('App_Usage_Type')['Usage_Hours'].sum().reset_index
app_usage_summary
```

```python
fig = px.pie(
    app_usage_summary,
    names='App_Usage_Type',       
    values='Usage_Hours',        
    title="Distribution of App Usage Hours by Type"
)

fig.show()
```
