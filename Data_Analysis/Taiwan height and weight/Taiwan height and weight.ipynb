# Analysis

## BMI, Weight , Height of different Age  

```python
sns.catplot(
    df,
    kind='bar',
    x=df['yr'],
    y=df['bmi'],
    hue='gender',
    height=7,
    aspect=2,
    
)

plt.title('bmi different Age')
plt.show()
```

```python
sns.catplot(
    df,
    kind='bar',
    x=df['yr'],
    y=df['height'],
    hue='gender',
    height=7,
    aspect=2,
)


plt.title('Height different Age')
plt.show()
```

```python
sns.catplot(
    df,
    kind='bar',
    x=df['yr'],
    y=df['weight'],
    hue='gender',
    height=7,
    aspect=2,
)

plt.title('Weight different Age')
plt.show()
```

## with year analysis

```python
sns.relplot(
    df,
    kind='line',
    x=df['yr'],
    y=df['bmi'],
    hue='gender',
    height=7,
    aspect=2,
)

plt.title('BMI with year')
plt.show()
```

```python
sns.relplot(
    df,
    kind='line',
    x=df['yr'],
    y=df['weight'],
    hue='gender',
    height=7,
    aspect=2,
)

plt.title('Weight with year')
plt.show()
```

```python
sns.relplot(
    df,
    kind='line',
    x=df['yr'],
    y=df['height'],
    hue='gender',
    height=7,
    aspect=2,
)

plt.title('height with year')
plt.show()
```
